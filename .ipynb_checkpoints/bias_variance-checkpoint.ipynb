```python
import pandas as pd
import numpy as np
import progressbar
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

import multiprocessing as mp
```

```python
sns.set()
```

# Sample data

```python
x=np.random.normal(0,1,(1000000,10))
w=np.random.normal(0,0.1,10)
h=np.random.normal(0,0.1,10)
```

### target

```python
def fnt(x):
    return (x.dot(w))**2+x.dot(h)+(x.dot(w))**4
```

```python
y=fnt(x)+np.random.normal(0,1,y.shape[0])
```

```python
indices=np.arange(x.shape[0])
np.random.shuffle(indices)
```

# Regressor model (decision tree)

```python
params={'criterion': 'mse',
 'max_depth': None,
 'max_features': None,
 'max_leaf_nodes': None,
 'min_impurity_decrease': 0.0,
 'min_impurity_split': None,
 'min_samples_leaf': 1,
 'min_samples_split': 2,
 'min_weight_fraction_leaf': 0.0,
 'random_state': None,
 'splitter': 'best'}
```

### sampling - training - predict 

```python
i=np.random.choice(indices[:10],1,replace=False)
x0=x[i]

max_features=[1,3,5,7,10]
models={}

def sampling(t):
    model, seed=t
    np.random.seed(seed)
    model.random_state=seed%11
    idx=np.random.choice(indices[10:],2*10**4,replace=False)
    train_x=x[idx]
    train_y=y[idx]
    model.fit(train_x,train_y)
    return model.predict(x0)
        
for f in progressbar.progressbar(max_features):
    params['max_features']=f
    models[f]={}
    for d in range(1,31):
        params['max_depth']=d
        
        with mp.Pool(mp.cpu_count()) as pool:
            models[f][d]=pool.map(sampling,[(DecisionTreeRegressor(**params),i) for i in range(100)])
```

# bias and variance calculation

```python
bias={}
var={}
for f in progressbar.progressbar(max_features):
    bias[f]={}
    var[f]={}
    for d in range(1,31):
        bias[f][d]=[]
        var[f][d]=[]
        y_pred=models[f][d][0]
        #predictions
        for m in models[f][d][1:]:
            y_pred=np.concatenate((y_pred,m),axis=0)
        bias[f][d]=y_pred.mean()-fnt(x0)
        var[f][d]=y_pred.var()
```

```python
df_bias={}
for f in bias:
    df_bias[f]=[]
    for _,value in bias[f].items():
        df_bias[f].append(value[0]**2)
df_bias['depth']=[d for d in bias[5]]
df_bias=pd.DataFrame.from_dict(df_bias)
```

```python
df_var={}
for f in var:
    df_var[f]=[]
    for _,value in var[f].items():
        df_var[f].append(value)
df_var['depth']=[d for d in var[5]]
df_var=pd.DataFrame.from_dict(df_var)
```

```python
df_mse=df_bias.iloc[:,:-1]+df_var.iloc[:,:-1]
```

```python
df_mse['depth']=[d for d in bias[5]]
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
plt.subplots_adjust(wspace=0.3,hspace=0.3)
for i,m in enumerate(max_features[:-1]):
    sns.lineplot(data=df_bias,x='depth',y=m,label='bias^2',ax=axes[i//2,i%2])
    sns.lineplot(data=df_var,x='depth',y=m,label='var',ax=axes[i//2,i%2])
    sns.lineplot(data=df_mse,x='depth',y=m,label='mse',ax=axes[i//2,i%2])
    plt.legend()
    axes[i//2,i%2].set_ylabel('error')
    axes[i//2,i%2].set_title('max_features={}'.format(m))
plt.savefig('bias_variance_multiple.png')
plt.show()
```

```python
m=10
plt.figure(figsize=(6,6))
sns.lineplot(data=df_bias,x='depth',y=m,label='bias^2')
sns.lineplot(data=df_var,x='depth',y=m,label='var')
sns.lineplot(data=df_mse,x='depth',y=m,label='mse')
plt.legend()
plt.ylabel('error')
plt.title('max_features={}'.format(m))
plt.savefig('bias_variance_maxfeatures.png')
plt.show()
```
