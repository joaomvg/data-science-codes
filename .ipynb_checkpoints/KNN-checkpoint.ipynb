```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import accuracy_score
sns.set()
```

```python
x=np.random.normal(0,10,(10**3,2))
```

```python
def y(x):
    norm=np.linalg.norm(x,axis=1)
    idx0=norm>10 
    idx1=(norm<=10) & (norm>5)
    idx2=norm<=5
    ypred=np.zeros(x.shape[0])
    ypred[idx0]=0
    ypred[idx1]=1
    ypred[idx2]=2
    
    return ypred
```

```python
ydata=y(x)
```

```python
data=np.concatenate([x,y(x).reshape(-1,1)],axis=1)
```

```python
df=pd.DataFrame(data=data,columns=['x','y','target'])
```

```python
df.target=df.target.astype('int32')
```

```python
sns.scatterplot(data=df,x='x',y='y',hue='target')
```

```python
class KNN:
    def __init__(self,k):
        self.k=k
        self.x=None
        self.y=None
        self.classes=None
        
    def fit(self,x,y):
        self.x=x
        classes=sorted(set(y))
        self.classes={a:b for b,a in enumerate(classes)}
        self.y=np.zeros((y.shape[0],len(classes)))
        
        for i,a in enumerate(y):
            j=self.classes[a]
            self.y[i][j]=1
    
    def __call__(self,x):
        if len(x.shape)==1:
            t=x.reshape(1,-1)
        else:
            t=x
        y_pred=np.zeros((t.shape[0],1))
        for i,z in enumerate(t):
            z=z.reshape(1,-1)
            norm=self.x-z
            norm=np.linalg.norm(norm,axis=1)
            args=np.argpartition(norm,self.k)[:self.k]
            ypred=self.y[args].mean(0)
            ypred=ypred.argmax()
            y_pred[i]=ypred
        
        return y_pred
```

```python
knn=KNN(5)
```

```python
knn.fit(df[['x','y']].values,df['target'].values)
```

```python
ypred=knn(x)
```

```python
ypred[0]
```

```python
accuracy_score(ypred,ydata)
```

```python
indices=np.arange(0,x.shape[0])
np.random.shuffle(indices)
l=int(0.2*x.shape[0])
xtrain=x[indices][:-l]
xtest=x[indices][-l:]

ytrain=ydata[indices][:-l]
ytest=ydata[indices][-l:]
```

```python
knn.fit(xtrain,ytrain)
```

```python
accuracy_score(ytest,knn(xtest))
```

# Iris dataset

```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
```

```python
iris=load_iris()
```

```python
features=iris['data']
target=iris['target']
```

```python
indices=np.arange(0,features.shape[0])
np.random.shuffle(indices)
l=int(0.2*features.shape[0])
xtrain=features[indices][:-l]
xtest=features[indices][-l:]

ytrain=target[indices][:-l]
ytest=target[indices][-l:]
```

```python
KNN=KNeighborsClassifier(n_neighbors=3)
```

```python
KNN.fit(xtrain,ytrain)
```

```python
knn.k=3
knn.fit(xtrain,ytrain)
```

```python
accuracy_score(ytest,knn(xtest))
```

```python
accuracy_score(ytest,KNN.predict(xtest))
```
