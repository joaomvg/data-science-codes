```python
class node:
    def __init__(self,tree=None,gamma=None):
        self.tree=tree
        self.gamma=gamma
        self.next=None
        
    def insert(self,tree,gamma):
        
        if self.next is None:
            self.next=node(tree,gamma)
        else:
            return self.next.insert(tree,gamma)
    
    def output(self,x):
        if self.tree is not None:
            out=self.tree.predict(x)
        if self.next is None:
            return out*self.gamma
        else:
            return out*self.gamma+self.next.output(x)
        
class GradBoost:
    def __init__(self,n_iter):
        self.n_iter=n_iter
        self.head=node()
        self.classes=None
        self.gamma0=None
    
    def __minima(self,h,F,y):
        g=-1
        g_prev=-1
        cl=self.classes[0]
        yr=y.copy()
        yr[y!=cl]=1
        yr[y==cl]=0
        
        not_converged=True
        i=0
        while not_converged:
            grad_dd=np.divide(h*h,(yr-F-g*h)*(yr-F-g*h)).sum()
            grad_d=np.divide(h,yr-F-g*h).sum()
            delta=-grad_d/grad_dd
            g+=delta
            i+=1
            if np.abs(g_prev-g)<0.001:
                not_converged=False
            if i>1000:
                break
            g_prev=g
        print(not_converged)
        return g
    
    def fit(self,x,y):
        self.classes=sorted(list(set(y)))
        self.dic={c:i for i,c in enumerate(self.classes)}
        cl=self.classes[0]
        gamma=(y==cl).sum()/y.shape[0]
        
        self.gamma0=gamma
        r=y.copy()
        r[y==cl]=0
        r[y!=cl]=1
        
        #1st STEP
        F=self.gamma0
        res=np.divide(1,F*np.ones(x.shape[0]))
        tree=DecisionTreeRegressor(max_depth=3)
        tree.fit(x,res)
        h=tree.predict(x)
        gamma=self.__minima(h,F,y)
        self.head.tree=tree
        self.head.gamma=gamma
        
        for i in range(1,self.n_iter):
            F=self.gamma0+self.head.output(x)
            res=np.divide(1,F)
            tree=DecisionTreeRegressor(max_depth=3)
            tree.fit(x,res)
            h=tree.predict(x)
            gamma=self.__minima(h,F,y)
            self.head.insert(tree,gamma)
            
    def predict(self,x):
        ypred=self.gamma0+self.head.output(x)
        cl=self.classes[0]
        return np.sign(ypred-0.5)*cl
    
    def predict_prob(self,x):
        ypred=self.gamma0+self.head.output(x)
        
        return ypred
```
